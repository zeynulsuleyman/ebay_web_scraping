```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd 


link = "https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1313&_nkw=men%27s+sports+shorts&_sacat=0"

mycontent = requests.get(link)

soup = BeautifulSoup(mycontent.content,'html.parser')

the_ul = soup.find('ul',attrs={'class':'srp-results srp-grid clearfix'})

titles = []
prices = []
shippings = []
urls = []
total_costs = []


for items in the_ul.findAll('li',attrs={'class':"s-item"}):
    title = items.find('h3',attrs={'class':"s-item__title"}).text
    price = items.find('span',attrs={'class':"s-item__price"}).text
    shipping = items.find('span',attrs={'class':"s-item__shipping s-item__logisticsCost"}).text
    url = items.find('a',href=True)['href']
    


    price = price[1:6]

   

    if shipping[0] == '+':
        shipping = shipping[2:6]
    else:
        shipping = '0'

    total_cost = float(price) + float(shipping)

    titles.append(title)
    prices.append(price)
    shippings.append(shipping)
    urls.append(url)
    total_costs.append(total_cost)

mylst = pd.DataFrame({'Product Name:':titles,'Price: ':prices,'Shipping:':shippings,'URL: ':urls,'Total Cost:': total_costs})
mylst.to_csv('prices.csv')
```
